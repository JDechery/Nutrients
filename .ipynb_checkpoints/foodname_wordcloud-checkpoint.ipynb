## USDA Nutrition Database Foodnames

Database contained nutritional values for packaged, unpackaged, processed, and unprocessed foods.
Food names are vary widely in detail:
- CLAWSON, CARNIVAL, WHITE STILTON WITH MANGO & GINGER, UPC: 735006000425
- Butter, salted

Let's take a look at the frequency of words across all foods compiled.
*NB* this dataset is incomplete. Because of API request limits, I'm still compiling the database from the [USDA.gov](https://ndb.nal.usda.gov/ndb/doc/index) website


```python
"""Import statements."""
import itertools
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import numpy as np
import re
from wordcloud import WordCloud
%matplotlib inline
```

```python
dbfile = 'F:/Data/nutrients_database.sqlite'
conn = sqlite3.connect(dbfile)

foods = pd.read_sql_query('SELECT * from food', conn)
foodNames = foods['name'].values

conn.close()
```

#### Regex to split full food name into individual words, ignoring special characters, punctuation, and numbers

```python
def splitNameIntoWords(nameList):
    """Split food name into list of words."""
    matchString = '[^\W\d]*'
    words = [list(filter(None, re.findall(matchString, name))) for name in nameList]
    return words
```

#### Some of the most common words are boring. Let's make sure to get rid of them

```python
wordsInFoods = splitNameIntoWords(foodNames)
allWords = list(itertools.chain(*wordsInFoods))
wordcounts = pd.Series(allWords).value_counts()
wordcounts.index = wordcounts.index.map(str.lower)
unfun_words = ['upc', 's', 'with', 'and', 'in', 'a', 'gtin', 'to', 'of']
wordcounts.drop(unfun_words, inplace=True)
```

#### Let's have some fun and make a [wordcloud](https://github.com/amueller/word_cloud)

I like masking it with an ellipse, rather than the default rectangle.
Next interation could be to take a stock image of some grocery items and embed it into the cloud using some more functionality provided by the package.

```python
# WordCloud wants mask as uint8 with disallowed pixels == 255
xx, yy = np.meshgrid(np.linspace(-10, 10, num=1600), np.linspace(-10, 10, num=1600))
mask = np.square(xx) + np.square(yy * 1.1) > 95
mask = mask.astype('uint8')*255
```

#### Using the top 150 words. Any more and it gets too difficult to read.

```python
wcloud = WordCloud(width=1600, height=1600, mask=mask, colormap='ocean').generate_from_frequencies(wordcounts[:150].to_dict())
plt.rcParams["figure.figsize"] = (12, 12)
plt.rcParams["axes.titlesize"] = 18
fig, ax = plt.subplots(1)
ax.imshow(wcloud)
ax.xaxis.set(ticks=[])
ax.yaxis.set(ticks=[])
plt.title('100 most common words at the grocery store')
# plt.savefig('most_common_words.png', dpi=300, bbox_inches='tight')
plt.show()
```
